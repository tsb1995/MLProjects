```python
import numpy as np
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn import metrics
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
def display_image(image):
    plt.imshow(image, interpolation='nearest')
    plt.show
```

```python
# Load the data

digits = load_digits()
data = digits.data
target = digits.target
images = digits.images
```

```python
# Get an idea of what the data looks like

print(data[14])
print(images[14])
display_image(images[18])
print("This image is classified as:", target[14])
```

```python
# Scale data and set up vectors/constants

data = scale(data)
x = data
y = target
k = 11 # Get number of categories
samples, features = data.shape
```

```python
# Setup classifier to cluster using KMeans

classifier = KMeans(n_clusters = k).fit(x)
y_pred = classifier.predict(x)
for i in range(len(y)):
    print("Predict:", y_pred[i], "Actual:", y[i])
```

```python
# Reduce to two dimensions for visualization

pca = PCA(n_components = 2, whiten = False)
X_pca = pca.fit_transform(x)
```

```python
# 2D graph of the data and its actual label

plt.figure(figsize=(12,12))

plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], color='red', alpha=0.5,label='0')
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], color='blue', alpha=0.5,label='1')
plt.scatter(X_pca[y==2, 0], X_pca[y==2, 1], color='green', alpha=0.5,label='2')
plt.scatter(X_pca[y==3, 0], X_pca[y==3, 1], color='black', alpha=0.5,label='3')
plt.scatter(X_pca[y==4, 0], X_pca[y==4, 1], color='khaki', alpha=0.5,label='4')
plt.scatter(X_pca[y==5, 0], X_pca[y==5, 1], color='yellow', alpha=0.5,label='5')
plt.scatter(X_pca[y==6, 0], X_pca[y==6, 1], color='turquoise', alpha=0.5,label='6')
plt.scatter(X_pca[y==7, 0], X_pca[y==7, 1], color='pink', alpha=0.5,label='7')
plt.scatter(X_pca[y==8, 0], X_pca[y==8, 1], color='moccasin', alpha=0.5,label='8')
plt.scatter(X_pca[y==9, 0], X_pca[y==9, 1], color='olive', alpha=0.5,label='9')
plt.scatter(X_pca[y==10, 0], X_pca[y==10, 1], color='coral', alpha=0.5,label='10')
plt.title("PCA")
plt.ylabel('Les coordonnees de Y')
plt.xlabel('Les coordonnees de X')
plt.legend()
plt.show()
```

```python
def create_mapping(k, y, y_pred):
    """
    Create a mapping between two cluster vectors based on mutual info score
    
    Arguments
    ---------
    
    k: number of clusters
    y: actual cluster vector
    y_pred: predicted cluster vector
    
    Returns
    --------
    mapping: vector of size k in which each value is the cluster in y_pred most closely associated with 
        the index-valued cluster in y
    """
    mapping = []
    for j in range(k):
        best = 0
        best_i = 0
        for i in range(k):
            dist = sklearn.metrics.mutual_info_score((y==j)*1, (y_pred==i)*1)
            if dist > best:
                best = dist
                best_i = i
        mapping.append(best_i) 
    return mapping
```

```python
# Create a mapping between y and y_pred so that the two graphs have properly corrosponding colors
# to give a visualization of how accurate our K-means algorithm was.

mapping = create_mapping(k, y, y_pred)
```

```python
plt.figure(figsize=(12,12))

plt.scatter(X_pca[y_pred==mapping[0], 0], X_pca[y_pred==mapping[0], 1], color='red', alpha=0.5,label='0')
plt.scatter(X_pca[y_pred==mapping[1], 0], X_pca[y_pred==mapping[1], 1], color='blue', alpha=0.5,label='1')
plt.scatter(X_pca[y_pred==mapping[2], 0], X_pca[y_pred==mapping[2], 1], color='green', alpha=0.5,label='2')
plt.scatter(X_pca[y_pred==mapping[3], 0], X_pca[y_pred==mapping[3], 1], color='black', alpha=0.5,label='3')
plt.scatter(X_pca[y_pred==mapping[4], 0], X_pca[y_pred==mapping[4], 1], color='khaki', alpha=0.5,label='4')
plt.scatter(X_pca[y_pred==mapping[5], 0], X_pca[y_pred==mapping[5], 1], color='yellow', alpha=0.5,label='5')
plt.scatter(X_pca[y_pred==mapping[6], 0], X_pca[y_pred==mapping[6], 1], color='turquoise', alpha=0.5,label='6')
plt.scatter(X_pca[y_pred==mapping[7], 0], X_pca[y_pred==mapping[7], 1], color='pink', alpha=0.5,label='7')
plt.scatter(X_pca[y_pred==mapping[8], 0], X_pca[y_pred==mapping[8], 1], color='moccasin', alpha=0.5,label='8')
plt.scatter(X_pca[y_pred==mapping[9], 0], X_pca[y_pred==mapping[9], 1], color='olive', alpha=0.5,label='9')
plt.scatter(X_pca[y_pred==mapping[10], 0], X_pca[y_pred==mapping[10], 1], color='coral', alpha=0.5,label='10')
plt.title("PCA")
plt.ylabel('Les coordonnees de Y')
plt.xlabel('Les coordonnees de X')
plt.legend()
plt.show()
```
