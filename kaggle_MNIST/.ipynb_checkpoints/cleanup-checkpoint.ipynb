```python
import numpy as np
import pandas as pd
import os
from keras.utils.np_utils import to_categorical # convert to one-hot-encoding
```

```python
train = pd.read_csv(os.path.join('data', 'train.csv'))
test = pd.read_csv(os.path.join('data', 'test.csv'))
```

```python
train.head()
```

```python
test.head()
```

```python
X_train = train.drop('label', axis=1)
y = train['label']
X_test = test
```

```python
print(X_train.shape)
print(y.shape)
print(X_test.shape)
```

```python
# Feature normalization (max greyscale value is 255)

X_train = X_train / 255.0
```

```python
# Reshape our input matrices to match shape of image (28x28) 
# with only one filter (would be 3 for RBG)
# and one hot encode our labels vector

X_train = X_train.values.reshape(-1, 28, 28, 1)
X_test = X_test.values.reshape(-1, 28, 28, 1)
y = to_categorical(y, num_classes=10)
```

```python
print(X_train.shape)
print(y.shape)
print(X_test.shape)
```

```python
np.save("data\X_train.csv", X_train)
np.save('data\y.csv', y)
np.save('data\X_test', X_test)
```
